```python
from sklearn.neighbors import KNeighborsClassifier #importing from scikit learn
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
```

n_neighbors=3 sets the value of k to 3

 scikit learn expect each objects to be vector/list, points or coordinates, see the X below
 y for classes, 0 and 1 we define 0 as blue, 1 as green. (y could show any attribute other than color but for simplicity we assume it's the color)

```python
X = [[0,0,0], [1,0,2], [2,0,3], [3,1,2], [1,1,1], [2,4,3], [3,1.5,2.5]]
y = [0, 0, 0, 1,1,1,0]
neigh3d = KNeighborsClassifier(n_neighbors=3)
```

the .fit method with uniform weights (default) which means all points are considered to have equal weights. 

```python
neigh3d.fit(X, y)
```

Note that we do a zip to pass all coordinates together to the scatter 

```python
list(zip(*X))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(*zip(*X), c='bbbgggb', s=100) #scatter wants a list of x and y we use zip to group them together so we use zip
ax.plot([1], [2], [3], '*r')# red star # passing the point [1,2,3] to the plot ([x], [y],[z])

```

```python
neigh3d.predict([[1,2,3]])
```

with k=3 predicted the red star to be class [0] which is blue
